# HAL Minimum Example

In this example, we add a few knowledge documents into a new empty memory and guide HAL to randomly sample a Gaussian distribution and plot it in a histogram.

````python
# install: pip install clelandlab-HAL
import HAL

# name will be used as source of new documents
HAL.init(name="Example Name", _config={
  "GEMINI_API_KEY": "your gemini API key", # can get from aistudio.google.com
  "MEMORY_DATA_PATH": "path/to/the/memory/data.gz", # if not existing, HAL will create one when saving memory
  "EXEC_IMPORT": "import time, os, sys, json, yaml, scipy, my_package\nimport numpy as np\nimport matplotlib.pyplot as plt\n"
})

# you can also save configs in a json file and load it with:
# HAL.init(_config="path/to/config.json")
````

````python
# simulate a package in the lab: my_package
import sys, types
import matplotlib.pyplot as plt

my_package = types.ModuleType("my_package")
my_package.__path__ = []  # Mark as a package
my_package.__file__ = "<in-memory>"

def _plot_histogram(data, bins=10):
    plt.hist(data, bins=bins)
    plt.show()
my_package.plot_histogram = _plot_histogram

sys.modules["my_package"] = my_package
````

````python
example_documents = ["""# Sample Distribution Experiment Plan

## Step 1

According to user's request, generate a list of numbers using the scipy package. Save the generated number array as STATE["numbers"]. If not specified, sample 1000 points.

## Step 2

Plot the generated numbers as a histogram using the `plot_histogram` function in `my_package`.
""", """# API Document: my_package.plot_histogram

Description: plot data as a histogram.

```python
my_package.plot_histogram(data, bins=10)
```

Argument:
- `data` (list): list of numbers
- `bins=10` (int): number of bins

No returns.
"""]
````

````python
# Add example documents to HAL's memory (knowledge base)
for d in example_documents:
    HAL.memorize(d)
````

````python
# start the HAL cycle. This will generate code in the next cell.
# check real-time status and sequence in the init cell output.
_ = HAL("Sample a Gaussian distribution with mean=0 and std=10")
````

````python
# continue without any user input
# This will generate another code cell
_ = HAL()
````

````python
# Save the memory (knowledge base) persistently
# It will be load automatically during init
HAL.memory.save()
````

````python
# Search the memory (knowledge base)
HAL.search("plot_histogram")
````
